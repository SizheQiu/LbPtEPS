```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ge_data = pd.read_csv('../data/exp_data/GSE143834_Lactobacillus_plantarum_gene_expression_pH_.csv')
ge_data = ge_data[['Gene name','con_1','con_2','pH5.5_1','pH5.5_2','pH5.0_1','pH5.0_1.1','eggNOG category']]
ge_data['eggNOG category'] = ge_data['eggNOG category'].fillna('')
ge_data.head()
```

```python
samples = ['con_1','con_2','pH5.5_1','pH5.5_2','pH5.0_1','pH5.0_1.1']
cluster_ge = {sample:{} for sample in samples}
for i in range(len(ge_data['Gene name'])):
    functions = list( list(ge_data['eggNOG category'])[i] )
    for func in functions:
        if func not in cluster_ge[samples[0]].keys():
            for sample in samples:
                cluster_ge[sample][func] = list(ge_data[sample])[i]
        else:
            for sample in samples:
                cluster_ge[sample][func] += list(ge_data[sample])[i]
```

```python
clsge_data = pd.DataFrame(cluster_ge)
clsge_data['eggNOG category'] = clsge_data.index
```

```python
clsge_data
```

```python
temp = pd.DataFrame(zip( [ clsge_data[sample][i] for sample in samples for i in range(len(clsge_data.index))],\
                       [sample for sample in samples for i in range(len(clsge_data.index))],\
                       list(clsge_data['eggNOG category'])*len(samples)),\
                    columns=['Expression','Sample','COG category'] )
g = sns.catplot(
    data=temp, kind="bar",
    x="Expression", y="COG category", hue="Sample",
    palette="bright", alpha=0.7, height=10
)
```

```python
ge_data = pd.read_csv('../data/exp_data/GSE143834_Lactobacillus_plantarum_gene_expression_pH_.csv')
ge_data['Annotation'] = ge_data['Annotation'].fillna('')
```

```python
eps_inds = []
for i in range(len(ge_data['Annotation'])):
    if 'exopolysaccharide' in list(ge_data['Annotation'])[i] or 'Exopolysaccharide' in list(ge_data['Annotation'])[i]:
        eps_inds.append(i)
eps_gedata = (ge_data.iloc[eps_inds]).reset_index().drop(['index'],axis=1)
```

```python
samples = ['con_1','con_2','pH5.5_1','pH5.5_2','pH5.0_1','pH5.0_1.1']
temp = pd.DataFrame(zip( [ eps_gedata[sample][i] for sample in samples for i in range(len(eps_gedata.index))],\
                       [sample for sample in samples for i in range(len(eps_gedata.index))],\
                       list(eps_gedata['Gene name'])*len(samples)),\
                    columns=['Expression','Sample','Gene name'] )
g = sns.catplot(
    data=temp, kind="bar",
    x="Gene name", y="Expression", hue="Sample",
    palette="bright", alpha=0.7, height=5,aspect=1.5 ,
)
```
