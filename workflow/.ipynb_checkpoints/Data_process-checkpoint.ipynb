# Growth with pH control

```python
growth_A = pd.read_excel('../data/Exp_data/Growth_pHcontrol_A.xlsx')
growth_B = pd.read_excel('../data/Exp_data/Growth_pHcontrol_B.xlsx')
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,4))
times = np.arange(0,16,2)
for x in [4.5, 5, 5.5, 6.5]:
    temp_pd = (growth_A[growth_A['pH']==x]).reset_index().drop(['index'],axis=1)
    y = np.mean( temp_pd.iloc[:,2:5].values, axis=1)
    yerr = np.std( temp_pd.iloc[:,2:5].values, axis=1)
    axs[0].errorbar( times, y, yerr=yerr, ecolor='black', linestyle='dotted',
        marker='o', markersize=3, elinewidth=2, label='pH'+str(x)+'(A)')

for x in [4.5, 5, 5.5, 6.5]:
    temp_pd = (growth_B[growth_B['pH']==x]).reset_index().drop(['index'],axis=1)
    y = np.mean( temp_pd.iloc[:,2:5].values, axis=1)
    yerr = np.std( temp_pd.iloc[:,2:5].values, axis=1)
    axs[1].errorbar( times, y, yerr=yerr, linestyle='dotted',  ecolor='black',
                   marker='o', markersize=3, elinewidth=2, label='pH'+str(x)+'(B)')
for ax in axs:
    ax.legend()
    ax.grid()
    ax.set_xlabel('Time(hr)')
    ax.set_ylabel('OD600')
plt.suptitle('Growth curves of Lactobacillus plantarum HMX2 under different pH values')
```

# Metabolomics

```python
raw_metab = pd.read_excel('../data/Exp_data/Metabolomics.xlsx')
raw_metab_free = pd.read_csv('../data/Exp_data/Metabolomics_freepH.csv')
```

```python
data=[] # convert unit to mM
for i in range( int(len(raw_metab_free.index)/3) ):
    temp = {'Time': int( raw_metab_free.iloc[3*i]['Sample'].replace('Freedom','').replace('h','').strip() )}
    for met in list( raw_metab_free.columns[2:] ):
        avg = (np.mean([raw_metab_free.iloc[3*i][met], \
            raw_metab_free.iloc[3*i+1][met], raw_metab_free.iloc[3*i+2][met]]))/1000
        std = np.std([raw_metab_free.iloc[3*i][met]/1000, \
            raw_metab_free.iloc[3*i+1][met]/1000, raw_metab_free.iloc[3*i+2][met]/1000])
        temp[met+'(avg)'] = avg
        temp[met+'(std)'] = std
    data.append(temp)
metab_free = pd.DataFrame(data)
# metab_free.to_csv('../data/Exp_data/Metabolomics_freepH_mM.csv',index=None)
```

```python
data=[] # convert unit to mM
for i in range( int(len(raw_metab.index)/3) ):
    temp = {'Sample':'pH'+raw_metab.iloc[3*i]['Sample'].split('-')[0].strip(),\
            'Time': int( raw_metab.iloc[3*i]['Sample'].split('-')[1].replace('h','').strip() ) }
    for met in list( raw_metab.columns[1:] ):
        avg = (np.mean([raw_metab.iloc[3*i][met], \
            raw_metab.iloc[3*i+1][met], raw_metab.iloc[3*i+2][met]]))/1000
        std = np.std([raw_metab.iloc[3*i][met]/1000, \
            raw_metab.iloc[3*i+1][met]/1000, raw_metab.iloc[3*i+2][met]/1000])
        temp[met+'(avg)'] = avg
        temp[met+'(std)'] = std
    data.append(temp)
metab = pd.DataFrame(data)
# metab.to_csv('../data/Exp_data/Metabolomics_mM.csv',index=None)
```

```python
def plot_met(met, metab_table, ax):
    metab = metab_table
    for cond in ['pH4.5', 'pH5', 'pH5.5', 'pH6.5']:
        temp_pd = (metab[metab['Sample']==cond]).reset_index().drop(['index'],axis=1)
        if cond == 'pH4.5':
            avg = list(temp_pd[met+'(avg)'])[1:5] + list(temp_pd[met+'(avg)'])[-2:]
            std = list(temp_pd[met+'(std)'])[1:5] + list(temp_pd[met+'(std)'])[-2:]
            Time = list(temp_pd['Time'])[1:5] + list(temp_pd['Time'])[-2:]
            ax.errorbar(Time, avg, yerr= std,linestyle='dotted', elinewidth=2, label=cond )
            ax.scatter(Time,avg, s=20)
        else:
            avg, std = temp_pd[met+'(avg)'], temp_pd[met+'(std)']
            ax.errorbar(temp_pd['Time'],avg, yerr= std,linestyle='dotted', elinewidth=2, label=cond )
            ax.scatter(temp_pd['Time'],avg, s=20)
    ax.set_ylabel(met+'(mM)')
    ax.set_xlabel('Time(hr)')
    ax.legend()
    ax.grid()
```

```python
cc2plot = ['Glucose', 'Lactic acid', 'Pyruvic acid', 'Succinic acid','Formic acid', 'Acetic acid']

fig, axs = plt.subplots(2,3,figsize=(18,8))
for i in range(len(cc2plot)):
    x,y = int(i/3), int(i%3)
    plot_met(cc2plot[i], metab, axs[x][y])
```

# Process proteomics dataset

```python
def save_clean_fasta( in_path, out_path ):
    temp_records = SeqIO.parse( in_path , "fasta")
    new_records = []
    for seq_record in temp_records:
        ID = str(seq_record.id).split('|')[1].strip()
        seq = str(seq_record.seq)
        new_records.append( SeqRecord(Seq(seq), id = ID, name="",description="") )
    SeqIO.write(new_records, out_path ,"fasta")
    print( len(new_records) )
```
